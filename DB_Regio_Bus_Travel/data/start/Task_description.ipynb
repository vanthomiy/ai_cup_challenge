# DB Regio Ticket Sales Prediction Challenge

## General Setting

Understanding mobility behaviour of citizens allows to optimize public transport services and contributes to the general goal of decarbonisation. This is especially true in rural areas with a low population density. In order to meet the citizen's demand, DB Regio offers two kinds of services: first regular bus routes with regular stops and a precise time schedule and second an on-demand bus service. The on-demand services can be booked online. It follows a predefined route but leaves out stations if there is no online request for the station. Reliably knowing the demand for a travel one day ahead offers large optimisation potential thus reducing costs and CO2 emissions.

## Your Task

In this setting, your task is to predict ticket sales for bus travel in the region of Passau, Bavaria, Germany.
It is split into two different kinds of traffic, regular routes and on-demand travel, making up two similar tasks.
The score for both sub-tasks is combined with equal weighting.

The data has been collected by DB Regio and is aggregated to ticket sales per hour per bus stop, which is also visualised in the figure below. 
There is additional data available for the regular routes, i.e. the route definitions and locations for most bus stops.

### Data

There are 12 different folds, each consisting of 3 weeks for training and 1 week for forecasting. 
This data is naturally sparse, both over the different bus stops as well as the time of day. For this reason, only the busiest e.g. 50 bus stops are selected for the forecast. For all of these bus stops, passenger data for all hours of the 365 days of 2019 is made available.
In essence, this means that there is a time series of length 365 minus reserved test data for each bus stop and hour, such as "Passau ZOB" and 8 o'clock. This results in e.g. $50 \cdot 24 = 1200$ time series with a total of $1200 \cdot 365 = 438000$ values.
For the regular route travel, only 350 values, up to 16.12.2019, are available.


The raw data that the prediction data is aggregated from is also made available for the weeks that are not in the test set.
Another data source comes from queries posted to the [_WOHIN DU WILLST_](https://wohin-du-willst.de/) service.
You can use additional external data, such as weather data, but no other data that directly reflects ticket sales.

### Evaluation


It is not critical to know exactly how many tickets are sold at each bus stop, but rather more important whether there is no, low or high demand.
For this reason, we formulate the task as a classification problem instead of regression: your model forecasts whether _0, 1, 2 or more than 2_ passengers alight at a bus stop.

The Evaluation metric is the macro-averaged F1 score. In a binary setting, e.g. looking only at the prediction for a single class $c$, it is calculated as follows for a prediction $\hat y$ and the true sale numbers $y$:

$F_1^c(y,\hat y) = 2 \dfrac{P(y,\hat y) \cdot R(y,\hat y)}{P(y,\hat y)+R(y,\hat y)}$ 

With Precision $P^c(y,\hat y) = \dfrac{tp_c}{tp_c + fp_c}$ and Recall $R^c(y,\hat y) = \dfrac{tp_c}{tp_c + fn_c}$

$tp_c$, $fp_c$ and $fn_c$ are the number true positive, false positive and false negative respectively for class $c$.

Macro averaging is taking the (unweighted) mean of the F1 score for each class. For classes $C$:

$F_1^{macro}(y, \hat y) = \frac{1}{|C|} \sum\limits_{c\in C} F_1^c(y, \hat y)$

The overall score for a task is calculated by averaging the scores over all folds.

The overall score for both tasks is their mean.

## Prediction Example: Regular Travel

We present a basic prediction example that illustrates the intended way to load, split and submit data.

We show this for the regular travel data, though for the on demand task the procedure is identical. 

### Data Loading

First we import pandas to load the regular routes training set.

We can create the train partition by removing the dummy rows that represent the test set samples.
We also digitize the target variable into categories for 0, 1, 2, and 3 ore more passengers.

```python
import pandas as pd
import numpy as np

regular_travel = pd.read_csv('regular_travel.csv', parse_dates=['date'])

# first we remove the dummy test set rows
regular_travel_train = regular_travel.dropna(subset = ['Passengers'])

y = np.digitize(regular_travel_train.Passengers, bins=(1, 2, 3))
```

### Dummy Baseline Estimator

We can devise a simple seasonal prediction baseline model, that forecasts the data from 7 days ago.

We also define a convenient wrapper to digitize predictions made by any estimator into the same bins as the target. This could also be implemented as part of a `sklearn.Pipeline`.

```python
from sklearn.base import BaseEstimator
import numpy as np

class SeasonalPredictor(BaseEstimator):
    def __init__(self, gap=7, date_col:str = 'date'):
        self.gap = gap
        self.date_col = date_col
        
    def fit(self, X, y):
        self.dates_ = X[self.date_col]
        self.y_ = np.array(y)
        return self
    
    def predict(self, X):
        pred_dates = (X[self.date_col] - pd.to_timedelta(self.gap, unit='d')).values
        return self.y_[np.where(np.isin(self.dates_, pred_dates))[0]]

    
class QuantizationWrapper(BaseEstimator):
    def __init__(self, estimator: BaseEstimator, bins=(1, 2, 3)):
        self.estimator = estimator
        self.bins=bins
        
    def fit(self, X, y):
        self.estimator.fit(X, y)
        return self
    
    def predict(self, X):
        predictions = self.estimator.predict(X)
        return np.digitize(predictions, bins=self.bins)
```

### Custom Dataset Splitter

Here we import a custom Dataset Splitter to generate validation and test spits.

For validation we use 2 week training folds followed by a 7 day forecast window.

For testing we use a 3 week training period before the 7 predictions.

```python
from ts_split import GroupedTimeSeriesSplit
```

### Validation Scores

We can test our model using a validation split.
This is also the validation schema that you should use for model selection and hyper-parameter tuning.

Seasonal forecasting works to some degree because there is a strong weekly trend in the data.

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score, make_scorer



scores = cross_val_score(estimator=QuantizationWrapper(SeasonalPredictor()), X=regular_travel_train, 
                         y=y, groups=regular_travel_train.date, scoring=make_scorer(f1_score, average='macro'),
                         cv = GroupedTimeSeriesSplit(train_window=14, test_window=7, train_gap = 7))

scores.mean(), '+-', scores.std()
```

### Test Set Predictions

Predictions for the test set can be generated by looping over the 4 week folds.

All submissions must contain the three index columns, i.e. `date`, `EZone` and `hour`, to prevent alignment errors in score calculation.

```python
tscv = GroupedTimeSeriesSplit(train_window=21, test_window=7, train_gap = 0)
est = QuantizationWrapper(SeasonalPredictor())

predictions = []
for train_ind, test_ind in tscv.split(regular_travel, y=regular_travel.Passengers, dates = regular_travel.date):
    est = est.fit(regular_travel.iloc[train_ind], regular_travel.Passengers.iloc[train_ind])
    prediction_df = regular_travel.iloc[test_ind].copy()
    prediction_df.Passengers = est.predict(regular_travel.iloc[test_ind])
    predictions.append(prediction_df)
    
predictions = pd.concat(predictions)
predictions.to_csv('regular_submission.csv', index=False)
predictions.head()
```

## Prediction Example: On-demand travel

We can do the same with the on-demand travel data.

### Data Loading and Validation Scores

```python
import pandas as pd
import numpy as np

on_demand_travel = pd.read_csv('on_demand_travel.csv', parse_dates=['date'])

# first we remove the dummy test set rows
on_demand_travel_train = on_demand_travel.dropna(subset = ['Passengers'])

y = np.digitize(on_demand_travel_train.Passengers, bins=(1, 2, 3))


scores = cross_val_score(estimator=QuantizationWrapper(SeasonalPredictor()), X=on_demand_travel_train, 
                         y=y, groups=on_demand_travel_train.date, scoring=make_scorer(f1_score, average='macro'),
                         cv = GroupedTimeSeriesSplit(train_window=14, test_window=7, train_gap = 7))

scores.mean(), '+-', scores.std()
```

### Test Set Predictions

```python
tscv = GroupedTimeSeriesSplit(train_window=21, test_window=7, train_gap = 0)
est = QuantizationWrapper(SeasonalPredictor())

predictions = []
for train_ind, test_ind in tscv.split(on_demand_travel, y=on_demand_travel.Passengers, 
                                      dates = on_demand_travel.date):
    est = est.fit(on_demand_travel.iloc[train_ind], on_demand_travel.Passengers.iloc[train_ind])
    prediction_df = on_demand_travel.iloc[test_ind].copy()
    prediction_df.Passengers = est.predict(on_demand_travel.iloc[test_ind])
    predictions.append(prediction_df)
    
predictions = pd.concat(predictions)
predictions.to_csv('on_demand_submission.csv', index=False)
predictions.head()
```
